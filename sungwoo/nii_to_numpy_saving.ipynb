```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nibabel as nib
import timeit
import scipy.io as sio

from MPC_tools import *

%matplotlib inline
```

```python
base_dir = '/media/cocoan-gpu1/wani_8T/data/SEMIC'
project_type = 'semic'
type1 = 'model02_FIR_SPM_SINGLE_TRIAL'
type2 = 'model02_Overall_FIR_SPM_SINGLE_TRIAL'
```

```python
dir_path_list, filenames_list = making_file_path(base_dir, project_type, type1, type2)
```

```python
print(dir_path_list.shape)
print(dir_path_list)
```

```python
full_path_list = making_full_path_list(dir_path_list, filenames_list)
```

```python
for i in range(dir_path_list.shape[0]):
    load_index = np.arange(i,i+1)

    data_num = np.arange(0,len(filenames_list[i]))

    X, Y = load_nii(project_type, load_index, full_path_list, data_num)

    print('\n')

    temp_file_name = os.path.split(os.path.dirname(dir_path_list[i]))[1] + '_' + os.path.split(dir_path_list[i])[1]
    save_base_dir = '/home/cocoan-gpu1/D_drive/MPC/data/SEMIC/numpy_data/'
    save_file_name = os.path.join(save_base_dir, temp_file_name)

    start = timeit.default_timer()
    np.save(save_file_name,X)

    print('\n')
    print('$$$ Saved {}.npy $$$'.format(temp_file_name))
    print('\n')

    stop = timeit.default_timer()
    check_time = stop-start
    now = datetime.datetime.now()
    print("###  Saving Duration Time  : {} minutes   {} seconds      ###".format(int(check_time)//60, int(check_time)%60))
    print("###  Excecuted Time : {} ###".format(now))
    print('\n')

    del X

print('$$$ Saving is finished! $$$')
```
